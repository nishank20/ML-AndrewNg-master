## one vs all logistic regression

```python
%reload_ext autoreload
%autoreload 2

import numpy as np

import sys
sys.path.append('..')

from helper import nn
from helper import logistic_regression as lr

from sklearn.metrics import classification_report
```

```python
raw_X, raw_y = nn.load_data('ex3data1.mat')
print(raw_X.shape)
print(raw_y.shape)
```

```python
# add intercept=1 for x0
X = np.insert(raw_X, 0, values=np.ones(raw_X.shape[0]), axis=1)
X.shape
```

```python
# y have 10 categories here. 1..10, they represent digit 0 as category 10 because matlab index start at 1
# I'll ditit 0, index 0 again
y_matrix = []

for k in range(1, 11):
    y_matrix.append((raw_y == k).astype(int))

# last one is k==10, it's digit 0, bring it to the first position
y_matrix = [y_matrix[-1]] + y_matrix[:-1]
y = np.array(y_matrix)

y.shape
```

```python
t0 = lr.logistic_regression(X, y[0])
```

```python
print(t0.shape)
y_pred = lr.predict(X, t0)
print('Accuracy={}'.format(np.mean(y[0] == y_pred)))
```

```python
k_theta = np.array([lr.logistic_regression(X, y[k]) for k in range(10)])
print(k_theta.shape)
```

## making prediction
* think about the shape of k_theta, now you are making $X\times\theta^T$
> $(5000, 401) \times (10, 401).T = (5000, 10)$
* after that, you run sigmoid to get probabilities and for each row, you find the highest prob as the answer

```python
prob_matrix = lr.sigmoid(X @ k_theta.T)
```

```python
np.set_printoptions(suppress=True)
prob_matrix
```

```python
y_pred = np.argmax(prob_matrix, axis=1)
```

```python
y_answer = raw_y.copy()
y_answer[y_answer==10] = 0
```

```python
print(classification_report(y_answer, y_pred))
```

## feed forward NN

```python
%reload_ext autoreload
%autoreload 2

import numpy as np

import sys
sys.path.append('..')

from helper import nn
from helper import logistic_regression as lr

from sklearn.metrics import classification_report
```

## model
<img style="float: left;" src="img/nn_model.png">

```python
theta1, theta2 = nn.load_weight('ex3weights.mat')

theta1.shape, theta2.shape
```

```python
X, y = nn.load_data('ex3data1.mat',transpose=False)

X = np.insert(X, 0, values=np.ones(X.shape[0]), axis=1)  # intercept

X.shape, y.shape
```

## feed forward prediction

```python
a1 = X
```

```python
z2 = a1 @ theta1.T # (5000, 401) @ (25,401).T = (5000, 25)
z2.shape
```

```python
z2 = np.insert(z2, 0, values=np.ones(z2.shape[0]), axis=1)
```

```python
a2 = lr.sigmoid(z2)
a2.shape
```

```python
z3 = a2 @ theta2.T
z3.shape
```

```python
a3 = lr.sigmoid(z3)
a3
```

```python
y_pred = np.argmax(a3, axis=1) + 1  # numpy is 0 base index, +1 for matlab convention
y_pred.shape
```

```python
print(classification_report(y_pred, y_pred))
```
